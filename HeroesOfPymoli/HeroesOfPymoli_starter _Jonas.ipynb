```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data
#uniqueSN= purchase_data.SN.unique
unique_users= pd.DataFrame(purchase_data['SN'].unique())
#count of unique users
unique_users_ct = pd.DataFrame(unique_users.count(),columns=['Total Players'])
unique_users_ct
```

```python
#Number of Unique Items	Average Price	Number of Purchases	Total Revenue
print(purchase_data.columns)

#number of unique items 
unique_items = purchase_data['Item Name'].nunique()

#number of Purchases
num_purchases =  purchase_data['Purchase ID'].count()

#Total Revenue
Total_rev = purchase_data.Price.sum() #why no axis or why is there error when adding the axis


Total_rev = pd.to_numeric(Total_rev)
num_purchases = pd.to_numeric(num_purchases)
unique_items = pd.to_numeric(unique_items)

#Average Price
average_price = (Total_rev/num_purchases)


#Create Dataframe
d = {'Number of Unique Items': [unique_items], 'Average Price': [average_price], 'num_purchases':[num_purchases], 'Total Revenue':[Total_rev]}
summary = pd.DataFrame(data=d)
summary['Average Price'] = summary['Average Price'].map('${:,.2f}'.format)
summary['Total Revenue'] = summary['Total Revenue'].map('${:,.2f}'.format)

summary.style.hide_index() #### efficent way of hiding this????
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
###########categorize and bin
bins = [0,9.90,14.90,19.90,24.90,29.90,34.90,39.90,99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]

#pull only the columns we need
purchasing_analysis = purchase_data.loc[:,["Gender","SN","Age","Item ID","Item Name","Price"]]
purchasing_analysis["Ages Binned"]= pd.cut(purchase_data["Age"], bins,labels =group_names)#, labels=group_names)
#Perform Count
item_count = purchasing_analysis.groupby('Ages Binned')['Item ID'].count()
pd.DataFrame(item_count)
#Average Purchase Price
avg_price = purchasing_analysis.groupby('Ages Binned')['Price'].mean()
pd.DataFrame(avg_price)
#Total Purchase Value
avg_total_Price = item_count*avg_price
pd.DataFrame(avg_total_Price)
#Avg Total Purchase per person
User_count = purchasing_analysis.groupby('Ages Binned')['Ages Binned'].count()
Avg_total_per_person = avg_total_Price/User_count

item_count.dtype
#join all into one frame
Purchasing_analysis_final = pd.DataFrame({"Purchase Count": User_count, "Average Purchase Price": avg_price,"Total Purchase Value": avg_total_Price, "Average Total Purchase per Person": Avg_total_per_person})
Purchasing_analysis_final.sort_index()
```
